## import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##  read data

```python
import pandas as pd
data=pd.read_csv("attendence.csv")
```

```python
data
```

```python
data.columns = data.columns.str.strip()
```

```python
data.columns
```

## top 5 most present and least present employees

```python

top_5 = data.sort_values(by='Present Days', ascending=False).head(5)
top_5
```

```python
plt.figure()
plt.bar(top_5['Name'], top_5['Present Days'])
plt.title("Top 5 Most Present Employees")
plt.xlabel("Name")
plt.ylabel("Present Days")
plt.xticks(rotation=45)
plt.show()
```

```python
top_5_least=data.sort_values(by='Present Days', ascending=True).head(5)
top_5_least
```

```python
plt.figure()
plt.bar(top5_least['Name'], top5_least['Present Days'])

plt.title("Top 5 Least Present Employees")
plt.xlabel("Name")
plt.ylabel("Present Days")

plt.xticks(rotation=45)
plt.show()
```

## average number of absent days

```python
data.columns
```

```python

average_absent = data['Absent Days'].mean()
average_absent = data['Absent Days'].mean()
print("Average absent days:", average_absent)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(['Average'], [average_absent])

plt.title("Average Absent Days")
plt.ylabel("Absent Days")

plt.show()
```

## employees who taken leave only 2 days

```python
leave_2_days = data[data['Total Leave'] == 2]
leave_2_days
```

```python

plt.figure()
plt.bar(leave_2_days['Name'], leave_2_days['Total Leave'])

plt.xlabel("Employee Name")
plt.ylabel("Total Leave Days")
plt.title("Employees with Exactly 2 Leave Days")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
never_absent = data[data['Absent Days'] == 0]
```

```python
never_absent
```

## top 5 employees with the most overtime (days).


```python
top_5_ot = data.sort_values(by='OT Days', ascending=False).head(1)

print(top_5_ot[['Name', 'OT Days']])
```

```python
import matplotlib.pyplot as plt

plt.bar(top_5_ot['Name'], top_5_ot['OT Days'])

plt.xlabel("Employee Name")
plt.ylabel("OT Days")
plt.title("Top 5 Employees with Most Overtime (Days)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Top 10 late comers (days)

## early going patterns (who leaves early most often).

```python
early_going_patterns = data.sort_values(by='Early Going Days', ascending=False).head(10)
```

```python
early_going_patterns 
```

```python
import matplotlib.pyplot as plt

plt.bar(early_going_patterns['Name'], early_going_patterns['Early Going Days'])
plt.xlabel("Employee Name")
plt.ylabel("Early Going Days")
plt.title("Top 10 Employees Who Leave Early Most Often")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Weekly off utilization (who actually worked on weekly offs).

```python
data.columns
```

```python

worked_weekly_off = data[data['Weekly Off Present'] > 0]


worked_weekly_off_sorted = worked_weekly_off.sort_values(
    by='Weekly Off Present', ascending=False
)


top_weekly_workers = worked_weekly_off_sorted[['Name', 'Weekly Off Present']]

print(top_weekly_workers.head(10))
```

```python
plt.figure(figsize=(10,5))
plt.bar(top_weekly_workers['Name'], top_weekly_workers['Weekly Off Present'])

plt.xlabel("Employee Name")
plt.ylabel("Weekly Off Days Worked")
plt.title("Top 10 Employees Working on Weekly Offs")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
